#### Creating custom business day


```python
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

us_business_days = CustomBusinessDay(calendar=USFederalHolidayCalendar())
```

Now we can use this as a frequency in our date range function

```python
date_range = pd.date_range(start="2021-07-01", end="2021-07-30", freq=us_business_days)
date_range
```

```python
df = pd.read_csv('aapl.csv', parse_dates=["Date"], index_col="Date")
df = df['2021-07']
df = df.asfreq(freq=us_business_days)
df
```

### Create our own business days Class

If we go to this [link](https://github.com/pandas-dev/pandas/blob/master/pandas/tseries/holiday.py), pandas USFederaLHoliday calendar class is present 

```python
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday

class AlgerianHolidayCalendar(AbstractHolidayCalendar):
    """
    Algerian Government Holiday Calendar based on rules specified by:
    https://publicholidays.africa/algeria/2021-dates/
    """

    rules = [
        Holiday("New Year's Day", month=1, day=1),
        Holiday(
            "Amazigh New Year",
            month=1,
            day=12
        ),
        Holiday("Labour Day", month=5, day=1),
        Holiday("Eid al-Fitr", month=5, day=13),
        Holiday("Eid al-Fitr Holiday", month=5, day=14),
        Holiday("Indenpendance Day", month=7, day=5),
        Holiday("Eid al-Adha", month=7, day=20),
        Holiday("Eid al-Adha Holiday", month=7, day=21),
        Holiday("Islamic New Year", month=8, day=9),
        Holiday("Ashura", month=8, day=18),
        Holiday("Prophet Muhammad's Birthday", month=10, day=18),
        Holiday("Revolution Day", month=11, day=1)
]
```

```python
df = pd.read_csv('aapl.csv', parse_dates=["Date"], index_col="Date")
df = df['2021-07']

algerian_business_days = CustomBusinessDay(calendar=AlgerianHolidayCalendar(), weekmask='Sun Mon Tue Wed Thu')
df.asfreq(freq=algerian_business_days)
```
