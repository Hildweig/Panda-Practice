## Concatenation, merging dataframes

### Concatenating two dataframes

````python
algeria_weather = pd.DataFrame({
    "city": ["Blida", "Algiers", "Souk'Ahras"],
    "temperature": [17, 19, 10],
    "humidity": [60, 70, 50]})
algeria_weather
````

````python
england_weather = pd.DataFrame({
    "city": ["London", "Manchester", "Ramsgate"],
    "temperature": [12, 10, 11],
    "humidity": [70, 69, 74]})
england_weather
````

````python
df = pd.concat([algeria_weather, england_weather], ignore_index=True)
"""
If we didn't do ignore_index = True we would have 0 1 2 0 1 2 (the original indexes), and that's why I ignored
the index, since it doesn't even matter
"""
df
````

### Concatenate with keys + locating a specific df

You can associate keys to each dataframe by adding a parameters that takes a table of strings for each dataframe, however you can't ignore the index in this case.
You can even retrieve the data of a specific key using "loc" on the dataframe

````python
df = pd.concat([algeria_weather, england_weather], ignore_index=False, keys=["Algeria", "England"])
df
````

````python
df.loc["Algeria"]
````

### Concatenate columns
Sometimes you would instead like to have the second data frame in the columns, not in the rows. Let's say you have a dataframe that contains only temperature data, and the second one contains humidity, and you want to concatenate them to have both temperature and humidity in the same table

````python
temperature_df = pd.DataFrame({
    "city": ["Blida", "Algiers", "Souk'Ahras"],
    "temperature": [17, 19, 10],
})

humidity_df = pd.DataFrame({
    "city": ["Blida", "Algiers", "Souk'Ahras"],
    "humidity": [60, 70, 50]
})

weather_df = pd.concat([temperature_df, humidity_df], axis=1)
weather_df
````

###Concatenation Using Index
Indexing helps us to locate the rows so that even if the order isn't right, we would still have the right order in concatenation











````python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
````

````python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
````

````python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
````

### Merging with inner join

````python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
````

````python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
````

````python
df3 = pd.merge(df1, df2, on="city")
df3
````

### Merging with outer join

````python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
````

````python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
````

````python
df3=pd.merge(df1,df2,on="city",how="inner")
df3
````

````python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
````

### Left outer join
There is also left join, and right join, where in left join we would take the intersection, plus what does not intersect from the first data frame, while for the right join, we take the intersection, and what does not intersect in the second dataframe

````python
df3 = pd.merge(df1, df2, on= "city", how="left")
df3
````

````python
df3 = pd.merge(df1, df2, on="city", how="right")
df3
````

Let's say you have common columns for both frames, you can explicitely supply the suffixes



````python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
````

````python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
````

````python
df3= pd.merge(df1,df2,on="city",how="outer")
""" 
This one is without supplying the suffix
"""
df3
````

````python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
""" 
This one is with the suffixes
"""
````
